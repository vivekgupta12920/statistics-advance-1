Q1: What is a random variable?

  A random variable is a mathematical function that assigns a numerical value to each outcome in the sample space of a random experiment. It translates qualitative outcomes (like "Heads" or "Tails") into quantitative data for analysis.

 Q2: What are the types of random variables?

   Discrete: Variables with countable values (e.g., number of cars).Continuous: Variables that can take any value within a range (e.g., precise height).


 Q3: Difference between discrete and continuous distributions?

  Discrete distributions map probabilities to specific, individual points (mass), while continuous distributions map probabilities to intervals of values (density). In continuous distributions, the probability of any single exact point is zero.
  
Q4: What are Probability Distribution Functions (PDF)?

  A PDF describes the likelihood of a continuous random variable falling within a particular range. The area under the PDF curve between two points represents the probability of the variable occurring in that interval.
  
 Q5: How do CDFs differ from PDFs?

  The PDF shows the density at a specific point, whereas the Cumulative Distribution Function (CDF) shows the probability that the variable is less than or equal to a certain value. The CDF is the integral of the PDF.
  
Q6: What is a discrete uniform distribution?

  A distribution where a finite number of values are equally likely to be observed (e.g., rolling a fair six-sided die; each side has a \(1/6\) probability).

 Q7: Key properties of a Bernoulli distribution?

  It models a single trial with only two possible outcomes: "Success" (1) with probability \(p\), and "Failure" (0) with probability \(1-p\).
  
 Q8: What is the binomial distribution?

  It represents the number of successes in a fixed number (\(n\)) of independent Bernoulli trials. It is used to calculate the probability of "k" successes in "n" attempts (e.g., 5 heads in 10 coin flips).
  
 Q9: What is the Poisson distribution?

  It models the number of times an event occurs in a fixed interval of time or space. It is applied in scenarios like counting the number of emails received per hour or customers arriving at a bank.
  
Q10: What is a continuous uniform distribution?

  A distribution where all intervals of the same length within the distribution's range are equally probable (e.g., a random number generator picking any decimal between 0 and 1).

 Q11: Characteristics of a normal distribution?

  It is bell-shaped, perfectly symmetrical around the mean, and defined by its mean (\(\mu \)) and standard deviation (\(\sigma \)). The mean, median, and mode are all equal.

Q12: What is the standard normal distribution?

  A special case of the normal distribution where the mean is 0 and the standard deviation is 1. It is used to standardize different datasets for comparison using Z-scores.

Q13: What is the Central Limit Theorem (CLT)?

  The CLT states that the distribution of sample means will approach a normal distribution as the sample size increases (typically \(n\ge 30\)), regardless of the original population's shape.

 Q14: How does CLT relate to the normal distribution?

  The CLT provides the theoretical justification for using normal distribution tools (like Z-tests) on non-normal populations, provided the sample size is large enough.

Q15: Application of Z-statistics in hypothesis testing?

  Z-statistics are used to determine if the difference between a sample mean and a population mean is statistically significant or just due to chance, assuming the population variance is known.

Q16: How do you calculate a Z-score and what does it represent?

  Calculated as \(Z=(X-\mu )/\sigma \).It represents how many standard deviations a data point is above or below the population mean.

 Q17: Point vs. Interval estimates?

  A point estimate is a single value (like the sample mean \(\={x}\)), while an interval estimate (like a Confidence Interval) provides a range of plausible values for the population parameter.

 Q18: Significance of confidence intervals?

  They provide a measure of uncertainty. A 95% confidence interval means that if we repeated the sampling 100 times, 95 of those intervals would contain the true population parameter.

Q19: Relationship between Z-score and confidence interval?

  The Z-score determines the width of the confidence interval. For a 95% confidence interval, a "critical Z-value" of 1.96 is used to set the boundaries from the mean.

 Q20: How are Z-scores used to compare different distributions?

  By converting different scales into a standard scale (number of standard deviations), Z-scores allow you to compare apples to oranges—for example, comparing a student's SAT score to their ACT score.


Q21: Assumptions for applying the CLT?

  Samples must be independent.Samples must be identically distributed.The sample size must be sufficiently large (usually \(n\ge 30\)).

 Q22: Concept of expected value?

  The expected value \(E[X]\) is the long-term average or "center of gravity" of a probability distribution. It is the sum of all possible values weighted by their probabilities.

Q2 : Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low, high = 1, 7 # 1 to 6 inclusive
x = np.arange(low, high)
pmf = randint.pmf(x, low, high)

plt.stem(x, pmf)
plt.title("Discrete Uniform PMF (Fair Die)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

Q3 : Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pmf(x, p):
    """Calculates PMF for Bernoulli distribution at x (0 or 1)."""
    if x == 1: return p
    elif x == 0: return 1 - p
    return 0

p = 0.3
print(f"Probability of Success (x=1): {bernoulli_pmf(1, p)}")
```

Q9: Write a Python function to calculate Z-scores from a dataset and plot them?

```python
def plot_z_scores(dataset):
    mean = np.mean(dataset)
    std = np.std(dataset)
    z_scores = [(x - mean) / std for x in dataset]

    plt.scatter(range(len(z_scores)), z_scores)
    plt.axhline(0, color='red', linestyle='--')
    plt.title("Z-Scores of Dataset")
    plt.show()

plot_z_scores([10, 12, 23, 23, 16, 23, 12, 20])
```

Q10:  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?

```python
# Exponential distribution (highly non-normal)
pop_data = np.random.exponential(scale=2, size=10000)
sample_means = []

for _ in range(1000):
    sample = np.random.choice(pop_data, size=50) # Sample size n=50
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, color='coral', edgecolor='black')
plt.title("CLT: Distribution of Sample Means (from Exponential)")
plt.show()
```

Q11 : Simulate multiple samples from a normal distribution and verify the Central Limit Theorem?

```python
import numpy as np
import matplotlib.pyplot as plt

def verify_clt(mu=50, sigma=10, sample_size=40, iterations=1000):
    means = [np.mean(np.random.normal(mu, sigma, sample_size)) for _ in range(iterations)]
    plt.hist(means, bins=30, color='skyblue', edgecolor='black', density=True)
    plt.title(f"Distribution of Sample Means (n={sample_size})")
    plt.show()

verify_clt()
```

Q12 : Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)?

```python
from scipy.stats import norm

def plot_standard_normal():
    x = np.linspace(-4, 4, 100)
    plt.plot(x, norm.pdf(x, 0, 1), color='red')
    plt.fill_between(x, norm.pdf(x, 0, 1), alpha=0.2, color='red')
    plt.title("Standard Normal Distribution (μ=0, σ=1)")
    plt.show()

plot_standard_normal()
```

Q16 : Create a confidence interval for a dataset using Python and interpret the result?

```python
import scipy.stats as st

data = np.random.normal(50, 10, 100)
ci = st.t.interval(0.95, len(data)-1, loc=np.mean(data), scale=st.sem(data))
print(f"95% Confidence Interval: {ci}")
# Interpretation: We are 95% confident the true population mean lies between these bounds.
```

Q18: Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu, sigma = 0, 1  # Mean and standard deviation
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf, color='blue', lw=2)
plt.fill_between(x, pdf, alpha=0.2, color='blue')
plt.title(f'Normal Distribution PDF (μ={mu}, σ={sigma})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

Q23:  Implement Z-score normalization on a dataset and visualize its transformation?

```python
def z_score_normalize(data):
    return (data - np.mean(data)) / np.std(data)

original_data = np.random.normal(50, 10, 1000)
normalized_data = z_score_normalize(original_data)

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].hist(original_data, bins=30, color='orange', alpha=0.7)
ax[0].set_title("Original Data (μ=50, σ=10)")
ax[1].hist(normalized_data, bins=30, color='green', alpha=0.7)
ax[1].set_title("Z-Score Normalized (μ=0, σ=1)")
plt.show()
```

Q24: Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
from scipy.stats import skew, kurtosis

def calculate_shape(data):
    s = skew(data)
    k = kurtosis(data) # Fisher’s definition (Normal = 0)
    return s, k

data = np.random.normal(0, 1, 5000)
s, k = calculate_shape(data)

print(f"Skewness: {s:.4f} (Near 0 = Symmetrical)")
print(f"Kurtosis: {k:.4f} (Near 0 = Normal Tailedness)")
```
